```python
from bs4 import BeautifulSoup
import urllib2
page1 = urllib2.urlopen('http://espn.go.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/qualified/false').read()
page2 = urllib2.urlopen('http://espn.go.com/nba/statistics/player/_/stat/assists/qualified/false').read()
soup1 = BeautifulSoup(page1)
soup2 = BeautifulSoup(page2)
print type(soup1), (soup2)
print soup1.prettify()[0:1000], soup2.prettify()[0:1000]
table_div1 = soup1.find(id='my-players-table')
table_div2 = soup2.find(id='my-players-table')
print table_div1.prettify()[0:4000]
print table_div2.prettify()[0:4000]
table1 = table_div1.find("table")
table2 = table_div2.find("table")
print table1
print table2
table_head1 = table1.find(attrs={"class":'colhead'})
table_head2 = table2.find(attrs={"class":'colhead'})
print table_head1.prettify()
print table_head2.prettify()
header_cols1 = table_head1.findAll('td')
header_cols2 = table_head2.findAll('td')
print header_cols1
print header_cols2
cols = []
for header_col in header_cols1:
    val = header_col.string
    if val != 'RK':
        cols.append(val)
    if val == 'PLAYER':
        cols.append('POSITION')
print cols
#
table_rows1 = table1.findAll('tr')
table_rows2 = table2.findAll('tr')
print table_rows1
print table_rows2
#
players_stats_dicts1 = []
players_stats_array1 = []
for row in table_rows1:
    if row.attrs['class'][0]=='colhead':
        continue
    player_stats = []
    row_cols = row.find_all('td')
    col_vals = []
    player_col = row_cols[1]
    player_name = player_col.find('a').string
    player_position = player_col.contents[1]
    player_position = player_position.split(' ')[1]
    player_stats.append(player_name)
    player_stats.append(player_position)
    for i in range(2, len(row_cols)):
        stat = row_cols[i].string
        player_stats.append(stat)
    players_stats_array1.append(player_stats)
    player_stats = dict(zip(cols, player_stats))
    players_stats_dicts1.append(player_stats)
print players_stats_dicts1[0:5]  
print players_stats_array1[0:5]
#
players_stats_dicts2 = []
players_stats_array2 = []
for row in table_rows2:
    if row.attrs['class'][0]=='colhead':
        continue
    player_stats = []
    row_cols = row.find_all('td')
    col_vals = []
    player_col = row_cols[1]
    player_name = player_col.find('a').string
    player_position = player_col.contents[1]
    player_position = player_position.split(' ')[1]
    player_stats.append(player_name)
    player_stats.append(player_position)
    for i in range(2, len(row_cols)):
        stat = row_cols[i].string
        player_stats.append(stat)
    players_stats_array2.append(player_stats)
    player_stats = dict(zip(cols, player_stats))
    players_stats_dicts2.append(player_stats)
print players_stats_dicts2[0:5]  
print players_stats_array2[0:5]
#
import pandas as pd
import numpy as np
df = pd.DataFrame.from_dict(players_stats_dicts1)
df = pd.DataFrame.from_dict(players_stats_dicts2)
pandas.DataFrame.join()
df.head()
#
```
