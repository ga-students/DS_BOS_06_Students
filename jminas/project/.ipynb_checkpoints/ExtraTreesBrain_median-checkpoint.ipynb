```python
# load dataset
from sklearn.datasets.base import Bunch

dataset = {}
for cope in copes:
    files = []
    notfound = []
    found = []
    for subject in cleandata.index:
        fname = get_fmri_fname(subject, cope)
        if os.path.exists(fname):
            files.append(fname)
            found.append(subject)
        else:
            notfound.append(subject)
    print '%s: #subjects: %d, missing: %s' % (cope, len(files), str(notfound))
    # resample files:
    newfiles = []
    for filename in files:
        path, name, ext = split_filename(filename)
        newname = os.path.join(tmp_dir + '/' + '/'.join(path.split('/')[-3:]), name + '_2mm' + ext)
        if not os.path.exists(os.path.dirname(newname)):
            os.makedirs(os.path.dirname(newname))
        if not os.path.exists(newname):
            Resample(in_file=filename, resampled_file=newname, voxel_size=(2,2,2)).run()
        newfiles.append(newname)
    
    data2use = cleandata.select(lambda x: x in found)
    
    dataset[cope] = Bunch(func=newfiles,
                          dvkeys=dvkeys,
                          ivkeys=ivkeys,
                          behav=data2use[ivkeys],
                          targets=data2use[dvkeys])
```

```python
# create persistent CV splits for parallel execution
def persist_cv_splits(X, Xb, y, n_iter=5, data_dir='./', name='data',
    suffix="_cv_%03d.pkl", test_size=0.2, random_state=None):
    """Materialize randomized train test splits of a dataset."""

    cv = StratifiedShuffleSplit(y, n_iter=n_iter, test_size=test_size, random_state=random_state)
    cv_split_filenames = []
    for ii, (train, test) in enumerate(cv):
        cv_split_filename = os.path.abspath(os.path.join(data_dir, name + suffix % ii))
        if not os.path.exists(cv_split_filename):
            cv_fold = (X[train], Xb[train], y[train], X[test], Xb[test], y[test])
            joblib.dump(cv_fold, cv_split_filename)
        cv_split_filenames.append(cv_split_filename)
    
    return cv_split_filenames

cv_split_names = {}
for cope in copes:
    cope_cv_split_names = {}
    X = fmri_masked[cope]
    Xb = np.array(dataset[cope].behav)
    for key in dataset[cope].dvkeys:
        y = 2 * (np.array(dataset[cope].targets[key]) > cleandata[key].median()) - 1
        fnames= persist_cv_splits(X, Xb, y, n_iter=200, data_dir=tmp_dir, name='%s_%s' % (cope, key), suffix="_cv_%03d.pkl",
                                  test_size=0.2, random_state=None)
        cope_cv_split_names[key] = fnames
    cv_split_names[cope] = cope_cv_split_names
```

```python
def compute_evaluation(cv_split_filename, model, return_weights, params=None):
    """Function executed by a worker to evaluate a model on a CV split"""
    # All module imports should be executed in the worker namespace
    import numpy as np
    from sklearn.externals import joblib
    from sklearn import metrics
    from copy import deepcopy
    
    X_train, Xb_train, y_train, X_test, Xb_test, y_test = joblib.load(
        cv_split_filename, mmap_mode='c')
    
    if params is not None:
        model.set_params(**params)

    model_brain = deepcopy(model)
    model_behav = deepcopy(model)
    model_brain.fit(X_train, y_train)
    model_behav.fit(Xb_train, y_train)
    
    active_behav = np.where(model_behav.feature_weights > 0)[0] #np.percentile(model_behav.feature_weights, 10))[0]
    active_brain = np.where(model_brain.feature_weights > 0)[0] #np.percentile(model_brain.feature_weights, 10))[0]
    print "active_brain: %d" % len(active_brain)
    print "active_behav: %d" % len(active_behav)
    Xcombined_train = np.hstack((Xb_train[:, active_behav], X_train[:, active_brain]))
    Xcombined_test = np.hstack((Xb_test[:, active_behav], X_test[:, active_brain]))
    
    model.max_features = None
    model.fit(Xcombined_train, y_train)
    
    brain_validation_score = metrics.f1_score(y_test, model_brain.predict(X_test))
    behav_validation_score = metrics.f1_score(y_test, model_behav.predict(Xb_test))
    combined_validation_score = metrics.f1_score(y_test, model.predict(Xcombined_test))
    validation_score = (brain_validation_score, behav_validation_score, combined_validation_score)
    
    # get the feature weights for prediction from brain data
    if return_weights:
        if hasattr(model_brain, 'feature_weights'):
            weights = model_brain.feature_weights
        elif hasattr(model_brain, 'feature_importances_'):
            weights = model_brain.feature_importances_
        active = np.where(weights > 0)[0]
        weights = weights[active]
    '''
    # behavioral data only
    model.fit(Xb_train, y_train)
    behav_validation_score = metrics.f1_score(y_test, model.predict(Xb_test))
    
    # brain and behavioral combined
    Xcomb_train = np.c_[X_train, Xb_train]
    Xcomb_test = np.c_[X_test, Xb_test]
    model.fit(Xcomb_train, y_train)
    comb_validation_score = metrics.f1_score(y_test, model.predict(Xcomb_test))

    validation_score = (brain_validation_score, behav_validation_score, comb_validation_score)
    '''
    
    if return_weights:
        return validation_score, active, weights
    
    return validation_score


from iterative_tree import IterativeExtraTreesClassifier 
clf = IterativeExtraTreesClassifier(n_estimators=500, max_depth=None, 
                                    max_features=0.2,
                                    min_samples_split=1, n_jobs=1)

# try normal ExtraTreesClassifier and iterative version
clf_params = [{'max_iterations':1}, {'max_iterations':10}]
```

```python
tasks_linear = []
tasks_copes = {}
for cope in copes:
    tasks_keys = {}
    for key in dataset[cope].dvkeys:
        key_tasks = []
        for fname in cv_split_names[cope][key]:
            param_tasks = []
            for params in clf_params[:1]:
                task = lb_view.apply(compute_evaluation, fname, clf, True, params)
                param_tasks.append(task)
                tasks_linear.append(task)
            key_tasks.append(param_tasks)
        tasks_keys[key] = key_tasks
    tasks_copes[cope] = tasks_keys

n_tasks = len(tasks_linear)
print 'Started %d tasks' % n_tasks
```

```python
while True:
    n_completed = np.sum([task.ready() for task in tasks_linear])
    print("Tasks completed: {0}%".format(100 * n_completed / float(n_tasks)))
    if n_completed == n_tasks:
        break
    time.sleep(60)
    
if kill_cluster_engines:
    client.shutdown()
```

```python
# retrieve results
coperesults = {}

for cope, cope_tasks in tasks_copes.iteritems():
    coperesults[cope] = {}
    for key, param_tasks in cope_tasks.iteritems():
        coperesults[cope][key] = {}
        this_scores = []
        this_weights = []
        for ii in range(len(param_tasks[0])):
            this_scores.append(np.array([task[ii].get()[0] for task in param_tasks]))
            avg_weights = np.zeros(n_features, dtype=np.float)
            for task in param_tasks:
                _, active, weights = task[ii].get()
                avg_weights[active] += weights
            avg_weights /= len(param_tasks)
            this_weights.append(avg_weights)
        coperesults[cope][key]['scores'] = this_scores
        coperesults[cope][key]['weights'] = this_weights
```
