```python
def classify(X1, X, y):
    results = [[], []]
    scores = []
    clfs = []
    for train, test in StratifiedKFold(y, 10): #StratifiedKFold(y, 8): #StratifiedShuffleSplit(y, n_iter=200, test_size=0.1):
        clf1 = ExtraTreesClassifier(n_estimators=1000, max_features='sqrt')
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])
        cc = 1.
        while(abs(cc) > 0.05):
            Xrandom = np.random.randn(len(train), 1)
            cc = np.corrcoef(Xrandom.flatten(), y[train])[0, 1]

        Xhat = np.concatenate((X[train], Xrandom), axis=1)
        clf.fit(Xhat, y[train])
        fi = clf.steps[-1][1].feature_importances_
        fi[fi <= fi[-1]] = 0
        fidx = np.nonzero(fi)[0]
        print "num features: %d" % len(fidx)
        for i in range(10):
            Xhat = np.concatenate((X[train][:, fidx], Xrandom), axis=1)
            clf.fit(Xhat, y[train])
            fi = clf.steps[-1][1].feature_importances_
            fi[fi <= fi[-1]] = 0
            fidx = fidx[np.nonzero(fi)[0]]
        print "num features: %d" % len(fidx)
        
        
        #fidx = range(X.shape[1])
        clf1 = ExtraTreesClassifier(n_estimators=2000, max_features='sqrt')
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])
        ytest = clf.fit(X[train][:, fidx], y[train]).predict(X[test][:, fidx])
        #print y[test], ytest
        results[0].extend(y[test])
        results[1].extend(ytest)
        scores.append(f1_score(y[test], ytest))
        importances = np.zeros(X.shape[1])
        importances[fidx] = clf.steps[-1][1].feature_importances_
        #importances[clf.steps[-2][1].get_support()] = clf.steps[-1][1].feature_importances_
        #importances[np.abs(clf.steps[-2][1].coef_.flatten()) > 0] = clf.steps[-1][1].feature_importances_
        clfs.append(importances)
    roc_auc = roc_auc_score(results[0], results[1])
    #print confusion_matrix(results[0], results[1])
    #print classification_report(results[0], results[1])

    norm_scores = np.array(scores).dot(np.array(clfs))/np.sum(scores)
    fw = np.zeros(fmri_masked.shape[1])
    fw[X1.keys().values.astype(int)] = norm_scores
    return roc_auc, fw
```

```python
def null_variance_score(y_true, y_predict):
    y_true = np.array(y_true)
    y_predict = np.array(y_predict)
    ynull = []
    for train, test in LeaveOneOut(len(y_true)):
        ynull.append(np.average(y_true[train]))
    return 1 - np.sum((y_true - y_predict)**2)/np.sum((np.array(ynull) - y_true)**2)
```

```python
def regress(X1, X, y):
    results = [[], []]
    scores = []
    clfs = []
    for train, test in KFold(len(y), 10): #StratifiedKFold(y, 8): #StratifiedShuffleSplit(y, n_iter=200, test_size=0.1):
        clf1 = ExtraTreesRegressor(n_estimators=1000, max_features='sqrt')
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])
        Xhat = np.concatenate((X[train], np.random.randn(len(train), 1)), axis=1)
        clf.fit(Xhat, y[train])
        fi = clf.steps[-1][1].feature_importances_
        fi[fi <= fi[-1]] = 0
        fidx = np.nonzero(fi)[0]
        #print len(fidx)
        #fidx = range(X.shape[1])
        clf1 = ExtraTreesRegressor(n_estimators=2000, max_features='sqrt')
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])
        ytest = clf.fit(X[train][:, fidx], y[train]).predict(X[test][:, fidx])
        #print y[test], ytest
        results[0].extend(y[test])
        results[1].extend(ytest)
        scores.append(null_variance_score(y[test], ytest))
        importances = np.zeros(X.shape[1])
        importances[fidx] = clf.steps[-1][1].feature_importances_
        #importances[clf.steps[-2][1].get_support()] = clf.steps[-1][1].feature_importances_
        #importances[np.abs(clf.steps[-2][1].coef_.flatten()) > 0] = clf.steps[-1][1].feature_importances_
        clfs.append(importances)
    nv_score = null_variance_score(results[0], results[1])

    norm_scores = np.array(scores).dot(np.array(clfs))/np.sum(scores)
    fw = np.zeros(fmri_masked.shape[1])
    fw[X1.keys().values.astype(int)] = norm_scores
    return nv_score, fw
```

```python
def regress(X1, X, y, nobehav=True):
    results = [[], []]
    scores = []
    clfs = []
    subs = []
    for train, test in KFold(len(y), 10): #StratifiedKFold(y, 8): #StratifiedShuffleSplit(y, n_iter=200, test_size=0.1):
        clf1 = ExtraTreesRegressor(n_estimators=1000, max_features='sqrt')
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])

        cc = 1.
        while(abs(cc) > 0.05):
            Xrandom = np.random.randn(len(train), 1)
            cc = np.corrcoef(Xrandom.flatten(), y[train])[0, 1]

        if nobehav:
            Xhat = np.concatenate((X[train], Xrandom), axis=1)
        else:
            Xhat = np.concatenate((X[train][:, :-2], Xrandom), axis=1)
        clf.fit(Xhat, y[train])
        fi = clf.steps[-1][1].feature_importances_
        fi[fi <= fi[-1]] = 0
        fidx = np.nonzero(fi)[0]
        for i in range(10):
            Xhat = np.concatenate((X[train][:, fidx], Xrandom), axis=1)
            clf.fit(Xhat, y[train])
            fi = clf.steps[-1][1].feature_importances_
            fi[fi <= fi[-1]] = 0
            fidx = fidx[np.nonzero(fi)[0]]
            
        print len(fidx)
        #fidx = range(X.shape[1])
        if not nobehav:
            behavidx = np.arange(2) + X.shape[1] - 2
            fidx = np.concatenate((fidx, behavidx))
        clf1 = ExtraTreesRegressor(n_estimators=2000, max_features=None)
        clf = Pipeline([("standardize", StandardScaler()),
                        #('select', SelectKBest(k=20000)),
                        #('select2', LinearSVC(C=1e-1, penalty="l1", dual=False)),
                        ("forest", clf1)
                        ])
        ytest = clf.fit(X[train][:, fidx], y[train]).predict(X[test][:, fidx])
        #print y[test], ytest
        results[0].extend(y[test])
        results[1].extend(ytest)
        scores.append(null_variance_score(y[test], ytest))
        importances = np.zeros(X.shape[1])
        if nobehav:
            importances[fidx] = clf.steps[-1][1].feature_importances_
        else:
            importances[fidx] = clf.steps[-1][1].feature_importances_[:-2]
        subs.extend(X1.index[test].tolist())
        #importances[clf.steps[-2][1].get_support()] = clf.steps[-1][1].feature_importances_
        #importances[np.abs(clf.steps[-2][1].coef_.flatten()) > 0] = clf.steps[-1][1].feature_importances_
        clfs.append(importances)
    nv_score = explained_variance_score(results[0], results[1])

    norm_scores = np.array(scores).dot(np.array(clfs))/np.sum(scores)
    fw = np.zeros(fmri_masked.shape[1])
    fw[X1.keys()[:-2].values.astype(int)] = norm_scores
    return nv_score, fw, pd.DataFrame(results[1], index=subs, columns=['LSAS']) 
```
